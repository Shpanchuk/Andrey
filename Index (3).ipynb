```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def interpolator ():
    r = int(input('Радиус разбиения: '))
    n = int(input('Количество точек: '))
    a = np.arange(-r, r, 0.01)  
    b = 1/(1 + a**2)
    x = []
    y = []
    x = np.linspace(-r, r, n)
    y = (1/(1 + 25*x**2))
    def lagr (x, y, f):
        g = 0
        for j in range(len(y)):
            l = 1
            for i in range(len(x)):
                if i != j:
                    l *= (f - x[i]) / (x[j] - x[i])
            g += y[j] * l
        return g
    x1= np.linspace(np.min(x), np.max(x), 100)
    y1 = [lagr(x, y, i) for i in x1]
    plt.plot(a,b, x1, y1)
    plt.grid()
    plt.legend(("f(x)", "Интерполяция f(x)"))
    plt.show()
    def cheb(arg, x, y):
        result = 0
        for i in range(len(x)):
            basis = 1
            for j in range(len(x)):
                if i == j: continue
                basis = (arg - x[j]) / (x[i] - x[j]) * basis
            result = y[i]*basis + result
        return result
    xr = np.arange(-1, 1, 0.0001)
    def f(x):
        return 1/(25*x**2 + 1)
    xcheb = [np.cos(np.pi*(2*k-1)/20) for k in range(1, 11)]
    ycheb = [f(_) for _ in xcheb]
    y2 = []
    y2 = [cheb(_, xcheb, ycheb) for _ in xr]
    plt.plot(xr, y2)
    return
```
