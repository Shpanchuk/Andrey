x"+w^2*x=k*x^3


```python
# Пусть x' = a, тогда a' + x*w^2 = k*x^3, a' = k*x^3 - x*w^2
import numpy as np
import matplotlib.pyplot as plt
x0 = int(input())
y0 = int(input())
a0 = int(input())
w = int(input())
k = int(input())
N = int(input())
h = 0.001
x = np.zeros(N)
x[0] = x0
y = np.zeros(N)
y[0] = y0
a = np.zeros(N)
a[0] = a0
for i in range (N - 1):
    y[i + 1] = y[i] + h
    a[i + 1] = a[i] + h * (k * x[i]**3 - x[i] * w**2)
    x[i + 1] = x[i] + h * a[i]
plt.plot(x,y)
plt.grid()
plt.show()
```
